```python
import os
import numpy as np
```

```python
data_path = "train_data"
```

```python
lables = os.listdir(data_path)
categorys = np.arange(len(lables))
categorys_dict = dict(zip(lables,categorys))

print(lables)
print(categorys_dict)
```

```python
import cv2

data = []
target = []

face_classifire = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

for lable in lables:
    imgs_path = os.path.join(data_path,lable)
    imgs = os.listdir(imgs_path)
    for img in imgs:
        img_path = os.path.join(imgs_path,img)
        img = cv2.imread(img_path)

        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

        faces = face_classifire.detectMultiScale(gray)
        for (x,y,w,h) in faces:
            # cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
            croped_face = gray[y:y+h,x:x+w]
            cv2.imshow("live",croped_face)
            print("yes:y , no:n : ")
            k = cv2.waitKey(0)
            if(k == 121):
                croped_face = cv2.resize(croped_face,(50,50))
                data.append(croped_face)
                target.append(categorys_dict[lable])
            cv2.destroyWindow("live")
cv2.destroyAllWindows()
```

```python
data = np.array(data)
target = np.array(target)

print(data.shape, target.shape)

data = data.reshape(data.shape[0],data.shape[1]*data.shape[2])
np.save('data', data)
np.save('target', target)
```
